## Objetivo principal
El objetivo principal de este trabajo es disponibilizar un Servicio de Recomendaciones para Apuestas Deportivas, el cual al integrarse con diferentes plataformas provea al usuario final, es decir al jugador, de una experiencia única y customizada. Teniendo como fin último la mejora en el rendimiento económico de la plataforma consumidora del servicio.

El recomendador deberá nutrirse de la história de los usuarios para identificar gustos y preferencias, además deberá ser capaz de detectar similutes entre grupos de jugadores, permitiendo servir sugerencias precisas en tiempo real. El servicio deberá exponer una interfaz de consulta y obtención de resultados de alto impacto. La intención es mejorar la experiencia del usuario, junto a la tasa promedio de jugadas diarias y al monto medio apostado.

Se plantea como objetivo deseable que el servicio sea capaz de adaptarse dinámicamente a cambios en el comportamiento de los usuarios. Es importante brindar sugerencias no solo basandonos en el historial del jugador sino también de interacciones recientes. Se considera la posibilidad de que el servicio se alimente también de interacciones del usuario con el frontend de la plataforma de juego y no solo de las apuestas realizadas.

## Antecedentes
Actualmente los sistemas de recomendaciones han experimentado un crecimiento en su aplicabilidad y en sus casos de uso, teniendo quizás en el entretenimiento, las redes sociales y el marketing sus exponentes mas claros y tangibles. La idea de un sistema de recomendación es identificar relaciones de similitud entre los usuarios y poder sugerirles ciertas acciones afines a sus gustos y preferencias pasadas.

Dentro del entretenimiento, las apuestas deportivas representan un mercado en constante expansión, el tamaño de mercado calculado al 2023 fué de U$D 242 billones. A medida que los países comienzan a legislar y brindar un marco legal para que las apuestas en linea se desenvuelvan de la mejor manera, la competencia entre compañias que brindan estas plataformas tambien crece, convirtiendose en fundamental diferenciarse del resto brindando experiencias cada vez mas personalizadas a los jugadores.

### Tipos de sistema de recomendaciones
- **Basados en conocimiento o heurística**: Es posiblemente el tipo de recomendador más simple que exista, se basa en un conjunto de reglas y presupuestos que fueron definiendose con el correr del tiempo, en función del conocimiento del domínio. Suelen ser útiles en etapas tempranas y primeras aproximaciones. Este enfoque sortea eficazmente el conocido problema del "arranque en frio".
- **Basados en contenido**: Esta estratégia interpreta la experiencia pasada del usuario y las ofertas disponibles para realizar sus sugerencias que mejor se ajusten a la historia del cliente.
- **Filtro colaborativo**: Este approach mas sofisticado explora los intereses de cada usuario, sus gustos y posbile **proximidad** o **afinidad** con otros usuarios de intereses similares para realizar sugerencias de impacto.

### Métricas a considerar
- Variación en la interacción del usuario con la plataforma (engagement).
- Impacto en el valor medio del ticket.
- Impacto en la cantidad tickets diarios.

## Origen de los datos
La plataforma Betcortex se basa en una **arquitectura distribuida**, entre los componentes principales se pueden destacar **SGA** (Sistema de Gestión de Apuestas) y **MSCO** (Microservcio de Cotizaciones). Siendo el primero el corazón de la plataforma, en él se parametriza cada aspecto del juego de acuerdo a las necesidades del operador, es además el encargado de llevar registro de cada movimiento realizado por los jugadores. Mientras que el servicio de cotizaciones funciona como concentrador de eventos apostables. Cabe destacar que la plataforma se puede integrar con multiples proveedores de eventos, por lo que es necesario una interfaz unificada para el manejo de los mismos. En este sentido, el MSCO es el sistema encargado de recibir y adaptar la información provista por servicios externos a eventos manejables por el resto de la infraestructura de Betcortex. Para comunicar los diferentes servicios se utilizan diversos medios, entre los que se encuentran: Http, Kafka, Rabbit MQ, Websocket, entre otros.

```
SELECT
    *
FROM GruposTicket gt
INNER JOIN Tickets t ON t.IdGrupoTicket = gt.IdGrupoTicket
INNER JOIN LineasTicket lt ON lt.IdTicket = t.IdTicket
INNER JOIN msco_LineasTicket msco ON msco.IdTicket = lt.IdTicket 
                                AND msco.IdOfertaEvento = lt.IdOfertaEvento
```

Vemos un describe de nuestra tabla:

| **Nombre**                 | **Tipo**                |
|----------------------------|-------------------------|
| gt.IdGrupoTicket           | Int32                   |
| gt.FechaAlta               | DateTime64(3)           |
| gt.IdPuntoVenta            | Nullable(Int32)         |
| gt.IdCajero                | Nullable(Int32)         |
| gt.IdCaja                  | Nullable(Int32)         |
| gt.NroUsoCaja              | Nullable(Int32)         |
| gt.IdUsuario               | Nullable(Int32)         |
| gt.IdOT                    | String                  |
| gt.FechaDelay              | Nullable(DateTime64(3)) |
| gt.FechaConfirma           | Nullable(DateTime64(3)) |
| gt.Monto                   | Decimal(10, 2)          |
| gt.CotizacionRef           | Decimal(10, 2)          |
| gt.CotizacionReal          | Decimal(10, 2)          |
| gt.Premio                  | Decimal(10, 2)          |
| gt.Estado                  | String                  |
| gt.EsGanador               | Nullable(String)        |
| gt.FechaImpresion          | Nullable(DateTime64(3)) |
| gt.NoImprimible            | String                  |
| gt.Hash                    | Nullable(String)        |
| gt.FechaEvaluacion         | Nullable(DateTime64(3)) |
| t.IdTicket                 | Int32                   |
| t.IdGrupoTicket            | Int32                   |
| t.Codigo                   | Nullable(String)        |
| t.Monto                    | Decimal(10, 2)          |
| t.Premio                   | Decimal(10, 2)          |
| t.Recibido                 | Nullable(String)        |
| lt.IdTicket                | Int32                   |
| lt.IdOfertaEvento          | Int32                   |
| lt.CotizacionRef           | Decimal(10, 2)          |
| lt.CotizacionReal          | Decimal(10, 2)          |
| lt.Estado                  | String                  |
| lt.Orden                   | Int8                    |
| msco.IdTicket              | Int32                   |
| msco.IdOfertaEvento        | Int32                   |
| msco.IdProveedor           | Int16                   |
| msco.IdEvento              | Int32                   |
| msco.IdPartido             | Int32                   |
| msco.IdTorneo              | Int32                   |
| msco.IdDeporte             | Int32                   |
| msco.IdModalidad           | Int32                   |
| msco.IdTipoModalidad       | String                  |
| msco.CountryCode           | String                  |
| msco.OfertaEvento          | String                  |
| msco.Localia               | Int8                    |
| msco.Especificador         | String                  |
| msco.Evento                | String                  |
| msco.Partido               | String                  |
| msco.Torneo                | String                  |
| msco.Deporte               | String                  |
| msco.Pais                  | String                  |
| msco.Modalidad             | String                  |
| msco.PuntosCliente         | Decimal(10, 2)          |
| msco.Ganador               | String                  |
| msco.FechaInicioPartido    | Nullable(DateTime64(3)) |
| msco.CantOfertas           | Int16                   |

Limpiamos nuestros datos y visualizamos:


### Descripción de columnas clave:
1. **gt.FechaAlta**: Fecha de Iniciado el Grupo (todos) los tickets o fecha de alta del grupo de tickets. Es decir, desde que se comienza a armar el ticket (Estado N: No Confirmado). Alta de la primera apuesta (línea de ticket)
2. **gt.IdUsuario**: Clave Primaria de la Tabla Usuarios. Corresponde al cliente (si lo tuviere)
3. **gt.Monto**: Monto que se apuesta. Cuando se crea el grupo, éste vale $0.
4. **gt.Estado**: Estado del Grupo de tickets: A: Anulado - N: No Confirmado - C: Confirmado/Vendido - P: Pendiente (de autorizar cuando los montos son grandes) - W: Pendientes de Acreditar (Wait)
5. **gt.EsGanador**: Indica si el ticket es o no ganador. S: Si - N: No - NULL: indeterminado (antes de saber los resultados)
6.  **msco.Pais**: País donde se lleva a cabo el Torneo.
7.  **msco.Deporte**: Deporte de Evento.
8.  **msco.Torneo**: Toreno en el que circunscribe el Partido.
9.  **msco.Partido**: Partido sobre el que se realiza la apuesta.
10. **msco.Evento**: Evento apostable sobre el que se realiza la apuesta.

## FODA: oportunidades y amenazas; fortalezas y debilidades.
#### Fortalezas
1. Conjunto de datos reales y gran cantidad de registros.
2. Infraestructura técnica adecuada (Clickhouse, Airflow).
3. Experiencia previa del equipo con análisis de datos.
4. Proyecto con impacto real en una empresa operativa.

#### Debilidades
1. Tiempo limitado para el desarrollo completo del motor de personalización.
2. Complejidad en la integración del motor con la plataforma existente.

#### Oportunidades
1. Creciente demanda de personalización en plataformas de apuestas.
2. Posibilidad de aplicar este motor en otras plataformas.
3. Crecimiento del mercado de apuestas.

#### Amenazas
1. Competencia de otras plataformas que ya implementan sistemas de personalización avanzados.
2. Cambios en las regulaciones locales que afecten el uso de datos o el mercado de apuestas.
3. Posibles problemas de confidencialidad y privacidad de datos que podrían requerir sanitización.

## Definición de roles y funciones del equipo
**Jorge Leodeg**: Responsable de la coordinación general del proyecto, se encargará de supervisar las distintas fases y asegurarse de que el equipo cumpla con los plazos establecidos. Además, trabajará en la redacción de las entregas parciales y la Memoria final del trabajo.

**Javier Eduardo Báez Galárraga**: Encargado del análisis de datos. Javier será responsable de explorar el dataset, entender las tendencias y diseñar el algoritmo de recomendaciones basado en la información histórica.

**Leandro Emanuel Gutiérrez**: Responsable de la obtención, manipulación y preprocesamiento de datos, además de evaluar el rendimiento del sistema de recomendaciones. Leandro asegurará que los datos estén listos para el entrenamiento del modelo y trabajará en la optimización de los resultados obtenidos.

## Hipótesis 
Para realizar este TFM nos hemos propuesto corroborar o refutar las siguientes hipótesis:

**Hipótesis 1**. La popularidad del deporte/evento influye en la efectividad de las recomendaciones. 
Hipótesis Nula (H₀): La popularidad del deporte o evento no tiene un efecto significativo en la efectividad de las recomendaciones.

Hipótesis Alternativa (H₁): La popularidad del deporte o evento tiene un efecto significativo en la efectividad de las recomendaciones.

Método de Validación: Correlación entre la popularidad del deporte/evento (número de apuestas) y la precisión o recall del sistema de recomendación.

**Hipótesis 2**: Las similitudes entre usuarios basadas en características demográficas mejoran la precisión

Hipótesis Nula(H₀): Las similitudes basadas en características demográficas no mejoran la precisión de las recomendaciones en comparación con modelos basados únicamente en comportamientos individuales.

Hipótesis Alternativa (H₁): Las similitudes basadas en características demográficas mejoran la precisión de las recomendaciones en comparación con modelos basados únicamente en comportamientos individuales.

Método de Validación: Comparar las métricas de desempeño de ambos métodos utilizando un conjunto de datos de prueba.

**Hipótesis 3**: Los modelos de filtro colaborativo son más efectivos con mayor cantidad de datos históricos

Hipótesis Nula (H₀): La cantidad de datos históricos disponibles no afecta significativamente la efectividad de los modelos de filtro colaborativo.

Hipótesis Alternativa (H₁): Un mayor volumen de datos históricos mejora significativamente la efectividad de los modelos de filtro colaborativo.

**Hipótesis 4**:  Un modelo híbrido mejora la precisión en comparación con modelos individuales.
 
Hipótesis Nula (H₀): Un modelo híbrido no mejora la precisión de las recomendaciones en comparación con modelos individuales.
 
Hipótesis Alternativa (H₁): Un modelo híbrido mejora la precisión de las recomendaciones en comparación con modelos individuales.
Método de Validación: Comparar las métricas de precisión y recall entre modelos híbridos y modelos basados en filtro colaborativo o basado en contenido.

Método de Validación: Submuestreo del dataset para entrenar y evaluar modelos con diferentes volúmenes de datos históricos.

**Hipótesis 5**. Los patrones de comportamiento recientes tienen un mayor impacto que el historial completo. 

Hipótesis Nula (H₀): Los patrones de comportamiento recientes no tienen un mayor impacto en la precisión de las recomendaciones en comparación con el historial completo del usuario.

Hipótesis Alternativa (H₁): Los patrones de comportamiento recientes tienen un mayor impacto en la precisión de las recomendaciones en comparación con el historial completo del usuario.

Método de Validación: Comparar modelos entrenados con datos recientes frente a modelos entrenados con el historial completo utilizando un conjunto de validación.

## Marco Teórico
Los sistemas de recomendación representan una herramienta para predecir las preferencias de los usuarios y ofrecerles sugerencias personalizadas. Cada uno de nosotros como usuarios, de manera recurrente interactuamos con plataformas digitales como Netflix, Amazon, Spotify o LinkedIn cuyo motor de recomendaciones es uno de los elementos principales de su arquitectura. 
En esencia, un sistema de recomendación es un algoritmo que, a partir de un conjunto de datos sobre las interacciones de los usuarios con un catálogo de ítems, busca identificar patrones y relaciones que permitan predecir qué ítems serán del agrado de un usuario en particular. Estos sistemas, al personalizar la experiencia del usuario, no solo incrementan la satisfacción del mismo, sino que también optimizan los procesos de negocio, como la venta de productos o la difusión de contenido.

### Modelos de recomendación
En esta sección, introduciremos las técnicas de recomendación de mayor uso en diferentes industrias. 

Popularidad o Ranking.
Se trata del modelo más sencillo conceptualmente y fácil de implementar.  Se basa en identificar la frecuencia de compra/consumo de un producto, es decir, las veces que dos o más ítems se compran o consultan conjuntamente. Las recomendaciones dependen exclusivamente de la frecuencia y suelen no tener ningún grado de personalización.

#### Filtrado colaborativo
Esta técnica se basa en la premisa de que si dos usuarios tienen gustos similares, es probable que también les gusten los mismos ítems. El filtrado colaborativo se divide en dos enfoques principales:
En este enfoque, se identifican a los usuarios más similares a un usuario objetivo y se recomiendan los ítems que estos usuarios han valorado positivamente. Por ejemplo, si a un usuario le gustan las películas de ciencia ficción, el sistema buscará a otros usuarios a los que también les guste este género y les recomendará otras películas de ciencia ficción que hayan valorado bien.
Uno de los ejemplos más representativos de filtrado colaborativo se encuentra en Amazon. La plataforma de comercio electrónico utiliza algoritmos de filtrado colaborativo para recomendar productos a sus usuarios basándose en sus historial de compras, productos vistos y valoraciones. Si un usuario compra un libro de ciencia ficción, es probable que Amazon le recomiende otros libros de ciencia ficción o libros de autores similares, ya que asume que otros usuarios que han comprado el mismo libro también están interesados en ese género.

#### Filtrado Basado en Contenido
El filtrado basado en contenido, por su parte, se centra en las características de los ítems. En lugar de buscar usuarios similares, este enfoque busca ítems similares a los que un usuario ya ha consumido. Para ello, se construye un perfil de contenido para cada ítem, el cual describe sus características más relevantes. Posteriormente, se compara el perfil de un usuario con los perfiles de los ítems para encontrar aquellos que sean más similares.
Netflix es un ejemplo de cómo el filtrado basado en contenido puede mejorar la experiencia del usuario. La plataforma analiza las características de las películas y series que un usuario ha visto (género, director, actores, sinopsis) para recomendarle contenido similar. Por ejemplo, si un usuario ha visto muchas películas de acción, Netflix le recomendará otras películas de acción con características similares.
Este enfoque permite a Netflix ofrecer recomendaciones personalizadas incluso para usuarios nuevos que aún no tienen un historial de visualización extenso.

#### Modelos Híbridos
Los modelos híbridos combinan las fortalezas del filtrado colaborativo y basado en contenido. Estos modelos pueden mejorar la precisión de las recomendaciones al aprovechar la información tanto de los usuarios como de los ítems.
Spotify combina tanto el filtrado colaborativo como el basado en contenido para ofrecer a sus usuarios una experiencia de escucha personalizada. El sistema analiza el historial de escucha de un usuario, las características de las canciones (género, artista, tempo) y las preferencias de otros usuarios con gustos similares para crear listas de reproducción personalizadas.
Este enfoque híbrido permite a Spotify ofrecer recomendaciones más precisas y diversificadas, ya que combina la capacidad del filtrado colaborativo para identificar usuarios con gustos similares con la capacidad del filtrado basado en contenido para encontrar canciones con características específicas.
Como otros ejemplos, se destacan:
•	YouTube: Utiliza algoritmos híbridos para recomendar videos basados en el historial de visualización, las búsquedas y las interacciones del usuario con otros usuarios. 
•	LinkedIn: Recomienda contactos, empleos y cursos basados en el perfil profesional del usuario y las conexiones de sus contactos.

#### Sistemas de Recomendación en el Sector de las Apuestas
El sector de las apuestas, tradicionalmente basado en la intuición y la experiencia, ha experimentado una transformación significativa gracias a la incorporación de tecnologías entre las cuales destacan los sistemas de recomendación. Estos sistemas permiten a las casas de apuestas ofrecer a sus usuarios experiencias más personalizadas y relevantes, aumentando así su satisfacción y fidelidad incrementando así el negocio. 
El alcance de las recomendaciones puede incluir: 
•	Recomendaciones de eventos: Al analizar el historial de apuestas de un usuario, sus equipos favoritos y los mercados en los que suele apostar, los sistemas pueden recomendarle eventos deportivos específicos que podrían ser de su interés.
•	Recomendaciones de mercados: Dentro de un evento deportivo, los sistemas pueden sugerir mercados de apuestas menos obvios pero potencialmente más rentables para el usuario, basados en sus preferencias y en las tendencias del mercado.
•	Recomendaciones de cuotas: Al comparar las cuotas ofrecidas por diferentes casas de apuestas, los sistemas pueden identificar aquellas que ofrecen el mejor valor para un usuario determinado.
•	Creación de mercados personalizados: Para usuarios más experimentados, los sistemas pueden permitir la creación de mercados personalizados, adaptándose a sus estrategias de apuestas específicas.

## Establecimiento de hitos
Dado la complejidad del proyecto se propone el siguiente cronograma, considerando un equipo de desarrollo de 3 personas:

<div align="center">
  <img src="planproyecto.jpg" alt="Plan de Proyecto">
</div>

## Estimación de Costos
A continuación, se presenta un modelo para estimar los costos del proyecto en función de los siguientes factores:
1.	Duración del proyecto: Según el cronograma desglosado.
2.	Número de miembros del equipo: 3 miembros.
3.	Coste por hora: Supongamos un costo promedio de € 25 por hora por desarrollador.
4.	Margen adicional (20%): Para cubrir imprevistos y ajustes.

### Estimación de Horas de Trabajo
Con base en el cronograma detallado, calculamos el número aproximado de horas requeridas para cada fase del proyecto:

<div align="center">
  <img src="tablacostes.jpg" alt="Plan de Proyecto">
</div>


### Cálculo del Costo Base
El costo base se calcula como:
                                Costo Base=Horas Totales×Costo por Hora 
                                Costo Base=1440×25=€ 36,000

### Margen Adicional para Imprevistos (20%)
Se agrega un margen adicional para cubrir imprevistos o ajustes durante el desarrollo:
                                Margen=Costo Base×0.20=36,000×0.20= € 7,200
                                Costo Total del Proyecto
                                Costo Total=Costo Base+Margen=36,000+ 7,200= € 43,200

### Retorno de la Inversión
La implementación de este sistema de recomendación puede generar un retorno de la inversión a través de:
        • Aumento de la retención de usuarios: Al ofrecer recomendaciones personalizadas, se incrementa la satisfacción del usuario y su probabilidad de volver a la plataforma.
        • Incremento del valor promedio de las apuestas: Las recomendaciones pueden llevar a los usuarios a realizar apuestas de mayor valor.
        • Aumento de la frecuencia de las apuestas: Al ofrecer sugerencias relevantes, se incentiva a los usuarios a realizar más apuestas.
        • Mejora de la imagen de marca: Una plataforma con un sistema de recomendación personalizado es percibida como más innovadora y centrada en el usuario.
        • Mayor eficiencia operativa: Automatización de tareas y optimización de recursos.


## Referencias

1. Mali, M., Mishra, D., & Vijayalaxmi, M. (2022). Benchmarking for Recommender System (MFRISE). 3C TIC Cuadernos de Desarrollo Aplicados a Las TIC. https://doi.org/10.17993/3ctic.2022.112.146-156

2. Recommendation System. (s/f). NVIDIA Data Science Glossary. Recuperado el 23 de noviembre de 2024, de https://www.nvidia.com/en-us/glossary/recommendation-system/

3. Steck, H., L. Baltrunas, E. Elahi, D. Liang, Y. Raimond, and J. Basilico. 2021. Deep learning for recommender systems: A netflix case study. AI Magazine 42: 7–18. https://doi.org/10.1609/aaai.12013.

4. B. Smith and G. Linden, "Two Decades of Recommender Systems at Amazon.com," in IEEE Internet Computing, vol. 21, no. 3, pp. 12-18, May-June 2017, doi: 10.1109/MIC.2017.72.

5. Bischof, B., & Yee, H. (2023). Building recommendation systems in python and Jax: Hands-on production systems at scale. O’Reilly Media.

6. Banik, R. (2018). Hands-On Recommendation Systems with Python: Start building powerful and personalized, recommendation engines with Python. Packt Publishing.

7. Lisandro, K., Shenjun, Z., & Javier, K. (2017). Beating the bookies with their own numbers - and how the online sports betting market is rigged. En arXiv [stat.AP]. http://arxiv.org/abs/1710.02824
