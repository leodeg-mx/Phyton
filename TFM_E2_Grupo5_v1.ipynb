
Vemos un describe de nuestra tabla:

| **Nombre**                 | **Tipo**                |
|----------------------------|-------------------------|
| gt.IdGrupoTicket           | Int32                   |
| gt.FechaAlta               | DateTime64(3)           |
| gt.IdPuntoVenta            | Nullable(Int32)         |
| gt.IdCajero                | Nullable(Int32)         |
| gt.IdCaja                  | Nullable(Int32)         |
| gt.NroUsoCaja              | Nullable(Int32)         |
| gt.IdUsuario               | Nullable(Int32)         |
| gt.IdOT                    | String                  |
| gt.FechaDelay              | Nullable(DateTime64(3)) |
| gt.FechaConfirma           | Nullable(DateTime64(3)) |
| gt.Monto                   | Decimal(10, 2)          |
| gt.CotizacionRef           | Decimal(10, 2)          |
| gt.CotizacionReal          | Decimal(10, 2)          |
| gt.Premio                  | Decimal(10, 2)          |
| gt.Estado                  | String                  |
| gt.EsGanador               | Nullable(String)        |
| gt.FechaImpresion          | Nullable(DateTime64(3)) |
| gt.NoImprimible            | String                  |
| gt.Hash                    | Nullable(String)        |
| gt.FechaEvaluacion         | Nullable(DateTime64(3)) |
| t.IdTicket                 | Int32                   |
| t.IdGrupoTicket            | Int32                   |
| t.Codigo                   | Nullable(String)        |
| t.Monto                    | Decimal(10, 2)          |
| t.Premio                   | Decimal(10, 2)          |
| t.Recibido                 | Nullable(String)        |
| lt.IdTicket                | Int32                   |
| lt.IdOfertaEvento          | Int32                   |
| lt.CotizacionRef           | Decimal(10, 2)          |
| lt.CotizacionReal          | Decimal(10, 2)          |
| lt.Estado                  | String                  |
| lt.Orden                   | Int8                    |
| msco.IdTicket              | Int32                   |
| msco.IdOfertaEvento        | Int32                   |
| msco.IdProveedor           | Int16                   |
| msco.IdEvento              | Int32                   |
| msco.IdPartido             | Int32                   |
| msco.IdTorneo              | Int32                   |
| msco.IdDeporte             | Int32                   |
| msco.IdModalidad           | Int32                   |
| msco.IdTipoModalidad       | String                  |
| msco.CountryCode           | String                  |
| msco.OfertaEvento          | String                  |
| msco.Localia               | Int8                    |
| msco.Especificador         | String                  |
| msco.Evento                | String                  |
| msco.Partido               | String                  |
| msco.Torneo                | String                  |
| msco.Deporte               | String                  |
| msco.Pais                  | String                  |
| msco.Modalidad             | String                  |
| msco.PuntosCliente         | Decimal(10, 2)          |
| msco.Ganador               | String                  |
| msco.FechaInicioPartido    | Nullable(DateTime64(3)) |
| msco.CantOfertas           | Int16                   |

Limpiamos nuestros datos y visualizamos:


### Descripcin de columnas clave:
1. **gt.FechaAlta**: Fecha de Iniciado el Grupo (todos) los tickets o fecha de alta del grupo de tickets. Es decir, desde que se comienza a armar el ticket (Estado N: No Confirmado). Alta de la primera apuesta (lnea de ticket)
2. **gt.IdUsuario**: Clave Primaria de la Tabla Usuarios. Corresponde al cliente (si lo tuviere)
3. **gt.Monto**: Monto que se apuesta. Cuando se crea el grupo, ste vale $0.
4. **gt.Estado**: Estado del Grupo de tickets: A: Anulado - N: No Confirmado - C: Confirmado/Vendido - P: Pendiente (de autorizar cuando los montos son grandes) - W: Pendientes de Acreditar (Wait)
5. **gt.EsGanador**: Indica si el ticket es o no ganador. S: Si - N: No - NULL: indeterminado (antes de saber los resultados)
6.  **msco.Pais**: Pas donde se lleva a cabo el Torneo.
7.  **msco.Deporte**: Deporte de Evento.
8.  **msco.Torneo**: Toreno en el que circunscribe el Partido.
9.  **msco.Partido**: Partido sobre el que se realiza la apuesta.
10. **msco.Evento**: Evento apostable sobre el que se realiza la apuesta.

## FODA: oportunidades y amenazas; fortalezas y debilidades.
#### Fortalezas
1. Conjunto de datos reales y gran cantidad de registros.
2. Infraestructura tcnica adecuada (Clickhouse, Airflow).
3. Experiencia previa del equipo con anlisis de datos.
4. Proyecto con impacto real en una empresa operativa.

#### Debilidades
1. Tiempo limitado para el desarrollo completo del motor de personalizacin.
2. Complejidad en la integracin del motor con la plataforma existente.

#### Oportunidades
1. Creciente demanda de personalizacin en plataformas de apuestas.
2. Posibilidad de aplicar este motor en otras plataformas.
3. Crecimiento del mercado de apuestas.

#### Amenazas
1. Competencia de otras plataformas que ya implementan sistemas de personalizacin avanzados.
2. Cambios en las regulaciones locales que afecten el uso de datos o el mercado de apuestas.
3. Posibles problemas de confidencialidad y privacidad de datos que podran requerir sanitizacin.

## Definicin de roles y funciones del equipo
**Jorge Leodeg**: Responsable de la coordinacin general del proyecto, se encargar de supervisar las distintas fases y asegurarse de que el equipo cumpla con los plazos establecidos. Adems, trabajar en la redaccin de las entregas parciales y la Memoria final del trabajo.

**Javier Eduardo Bez Galrraga**: Encargado del anlisis de datos. Javier ser responsable de explorar el dataset, entender las tendencias y disear el algoritmo de recomendaciones basado en la informacin histrica.

**Leandro Emanuel Gutirrez**: Responsable de la obtencin, manipulacin y preprocesamiento de datos, adems de evaluar el rendimiento del sistema de recomendaciones. Leandro asegurar que los datos estn listos para el entrenamiento del modelo y trabajar en la optimizacin de los resultados obtenidos.

## Hiptesis 
Para realizar este TFM nos hemos propuesto corroborar o refutar las siguientes hiptesis:

**Hiptesis 1**. La popularidad del deporte/evento influye en la efectividad de las recomendaciones. 
Hiptesis Nula (H): La popularidad del deporte o evento no tiene un efecto significativo en la efectividad de las recomendaciones.

Hiptesis Alternativa (H): La popularidad del deporte o evento tiene un efecto significativo en la efectividad de las recomendaciones.

Mtodo de Validacin: Correlacin entre la popularidad del deporte/evento (nmero de apuestas) y la precisin o recall del sistema de recomendacin.

**Hiptesis 2**: Las similitudes entre usuarios basadas en caractersticas demogrficas mejoran la precisin

Hiptesis Nula(H): Las similitudes basadas en caractersticas demogrficas no mejoran la precisin de las recomendaciones en comparacin con modelos basados nicamente en comportamientos individuales.

Hiptesis Alternativa (H): Las similitudes basadas en caractersticas demogrficas mejoran la precisin de las recomendaciones en comparacin con modelos basados nicamente en comportamientos individuales.

Mtodo de Validacin: Comparar las mtricas de desempeo de ambos mtodos utilizando un conjunto de datos de prueba.

**Hiptesis 3**: Los modelos de filtro colaborativo son ms efectivos con mayor cantidad de datos histricos

Hiptesis Nula (H): La cantidad de datos histricos disponibles no afecta significativamente la efectividad de los modelos de filtro colaborativo.

Hiptesis Alternativa (H): Un mayor volumen de datos histricos mejora significativamente la efectividad de los modelos de filtro colaborativo.

**Hiptesis 4**:  Un modelo hbrido mejora la precisin en comparacin con modelos individuales.
 
Hiptesis Nula (H): Un modelo hbrido no mejora la precisin de las recomendaciones en comparacin con modelos individuales.
 
Hiptesis Alternativa (H): Un modelo hbrido mejora la precisin de las recomendaciones en comparacin con modelos individuales.
Mtodo de Validacin: Comparar las mtricas de precisin y recall entre modelos hbridos y modelos basados en filtro colaborativo o basado en contenido.

Mtodo de Validacin: Submuestreo del dataset para entrenar y evaluar modelos con diferentes volmenes de datos histricos.

**Hiptesis 5**. Los patrones de comportamiento recientes tienen un mayor impacto que el historial completo. 

Hiptesis Nula (H): Los patrones de comportamiento recientes no tienen un mayor impacto en la precisin de las recomendaciones en comparacin con el historial completo del usuario.

Hiptesis Alternativa (H): Los patrones de comportamiento recientes tienen un mayor impacto en la precisin de las recomendaciones en comparacin con el historial completo del usuario.

Mtodo de Validacin: Comparar modelos entrenados con datos recientes frente a modelos entrenados con el historial completo utilizando un conjunto de validacin.

## Marco Terico
Los sistemas de recomendacin representan una herramienta para predecir las preferencias de los usuarios y ofrecerles sugerencias personalizadas. Cada uno de nosotros como usuarios, de manera recurrente interactuamos con plataformas digitales como Netflix, Amazon, Spotify o LinkedIn cuyo motor de recomendaciones es uno de los elementos principales de su arquitectura. 
En esencia, un sistema de recomendacin es un algoritmo que, a partir de un conjunto de datos sobre las interacciones de los usuarios con un catlogo de tems, busca identificar patrones y relaciones que permitan predecir qu tems sern del agrado de un usuario en particular. Estos sistemas, al personalizar la experiencia del usuario, no solo incrementan la satisfaccin del mismo, sino que tambin optimizan los procesos de negocio, como la venta de productos o la difusin de contenido.

### Modelos de recomendacin
En esta seccin, introduciremos las tcnicas de recomendacin de mayor uso en diferentes industrias. 

Popularidad o Ranking.
Se trata del modelo ms sencillo conceptualmente y fcil de implementar.  Se basa en identificar la frecuencia de compra/consumo de un producto, es decir, las veces que dos o ms tems se compran o consultan conjuntamente. Las recomendaciones dependen exclusivamente de la frecuencia y suelen no tener ningn grado de personalizacin.

#### Filtrado colaborativo
Esta tcnica se basa en la premisa de que si dos usuarios tienen gustos similares, es probable que tambin les gusten los mismos tems. El filtrado colaborativo se divide en dos enfoques principales:
En este enfoque, se identifican a los usuarios ms similares a un usuario objetivo y se recomiendan los tems que estos usuarios han valorado positivamente. Por ejemplo, si a un usuario le gustan las pelculas de ciencia ficcin, el sistema buscar a otros usuarios a los que tambin les guste este gnero y les recomendar otras pelculas de ciencia ficcin que hayan valorado bien.
Uno de los ejemplos ms representativos de filtrado colaborativo se encuentra en Amazon. La plataforma de comercio electrnico utiliza algoritmos de filtrado colaborativo para recomendar productos a sus usuarios basndose en sus historial de compras, productos vistos y valoraciones. Si un usuario compra un libro de ciencia ficcin, es probable que Amazon le recomiende otros libros de ciencia ficcin o libros de autores similares, ya que asume que otros usuarios que han comprado el mismo libro tambin estn interesados en ese gnero.

#### Filtrado Basado en Contenido
El filtrado basado en contenido, por su parte, se centra en las caractersticas de los tems. En lugar de buscar usuarios similares, este enfoque busca tems similares a los que un usuario ya ha consumido. Para ello, se construye un perfil de contenido para cada tem, el cual describe sus caractersticas ms relevantes. Posteriormente, se compara el perfil de un usuario con los perfiles de los tems para encontrar aquellos que sean ms similares.
Netflix es un ejemplo de cmo el filtrado basado en contenido puede mejorar la experiencia del usuario. La plataforma analiza las caractersticas de las pelculas y series que un usuario ha visto (gnero, director, actores, sinopsis) para recomendarle contenido similar. Por ejemplo, si un usuario ha visto muchas pelculas de accin, Netflix le recomendar otras pelculas de accin con caractersticas similares.
Este enfoque permite a Netflix ofrecer recomendaciones personalizadas incluso para usuarios nuevos que an no tienen un historial de visualizacin extenso.

#### Modelos Hbridos
Los modelos hbridos combinan las fortalezas del filtrado colaborativo y basado en contenido. Estos modelos pueden mejorar la precisin de las recomendaciones al aprovechar la informacin tanto de los usuarios como de los tems.
Spotify combina tanto el filtrado colaborativo como el basado en contenido para ofrecer a sus usuarios una experiencia de escucha personalizada. El sistema analiza el historial de escucha de un usuario, las caractersticas de las canciones (gnero, artista, tempo) y las preferencias de otros usuarios con gustos similares para crear listas de reproduccin personalizadas.
Este enfoque hbrido permite a Spotify ofrecer recomendaciones ms precisas y diversificadas, ya que combina la capacidad del filtrado colaborativo para identificar usuarios con gustos similares con la capacidad del filtrado basado en contenido para encontrar canciones con caractersticas especficas.
Como otros ejemplos, se destacan:
	YouTube: Utiliza algoritmos hbridos para recomendar videos basados en el historial de visualizacin, las bsquedas y las interacciones del usuario con otros usuarios. 
	LinkedIn: Recomienda contactos, empleos y cursos basados en el perfil profesional del usuario y las conexiones de sus contactos.

## Sistemas de Recomendacin en el Sector de las Apuestas
El sector de las apuestas, tradicionalmente basado en la intuicin y la experiencia, ha experimentado una transformacin significativa gracias a la incorporacin de tecnologas entre las cuales destacan los sistemas de recomendacin. Estos sistemas permiten a las casas de apuestas ofrecer a sus usuarios experiencias ms personalizadas y relevantes, aumentando as su satisfaccin y fidelidad incrementando as el negocio. 
El alcance de las recomendaciones puede incluir: 
	Recomendaciones de eventos: Al analizar el historial de apuestas de un usuario, sus equipos favoritos y los mercados en los que suele apostar, los sistemas pueden recomendarle eventos deportivos especficos que podran ser de su inters.
	Recomendaciones de mercados: Dentro de un evento deportivo, los sistemas pueden sugerir mercados de apuestas menos obvios pero potencialmente ms rentables para el usuario, basados en sus preferencias y en las tendencias del mercado.
	Recomendaciones de cuotas: Al comparar las cuotas ofrecidas por diferentes casas de apuestas, los sistemas pueden identificar aquellas que ofrecen el mejor valor para un usuario determinado.
	Creacin de mercados personalizados: Para usuarios ms experimentados, los sistemas pueden permitir la creacin de mercados personalizados, adaptndose a sus estrategias de apuestas especficas.

## Establecimiento de hitos
Dado la complejidad del proyecto se propone el siguiente cronograma, considerando un equipo de desarrollo de 3 personas:



### Estimacin de Costos
A continuacin, se presenta un modelo para estimar los costos del proyecto en funcin de los siguientes factores:
1.	Duracin del proyecto: Segn el cronograma desglosado.
2.	Nmero de miembros del equipo: 3 miembros.
3.	Coste por hora: Supongamos un costo promedio de  25 por hora por desarrollador.
4.	Margen adicional (20%): Para cubrir imprevistos y ajustes.

### Estimacin de Horas de Trabajo
Con base en el cronograma detallado, calculamos el nmero aproximado de horas requeridas para cada fase del proyecto:




### Clculo del Costo Base
El costo base se calcula como:
                                Costo Base=Horas TotalesCosto por Hora 
                                Costo Base=144025= 36,000

### Margen Adicional para Imprevistos (20%)
Se agrega un margen adicional para cubrir imprevistos o ajustes durante el desarrollo:
                                Margen=Costo Base0.20=36,0000.20=  7,200
                                Costo Total del Proyecto
                                Costo Total=Costo Base+Margen=36,000+ 7,200=  43,200

### Retorno de la Inversin
La implementacin de este sistema de recomendacin puede generar un retorno de la inversin a travs de:
         Aumento de la retencin de usuarios: Al ofrecer recomendaciones personalizadas, se incrementa la satisfaccin del usuario y su probabilidad de volver a la plataforma.
         Incremento del valor promedio de las apuestas: Las recomendaciones pueden llevar a los usuarios a realizar apuestas de mayor valor.
         Aumento de la frecuencia de las apuestas: Al ofrecer sugerencias relevantes, se incentiva a los usuarios a realizar ms apuestas.
         Mejora de la imagen de marca: Una plataforma con un sistema de recomendacin personalizado es percibida como ms innovadora y centrada en el usuario.
         Mayor eficiencia operativa: Automatizacin de tareas y optimizacin de recursos.


## Referencias

1. Mali, M., Mishra, D., & Vijayalaxmi, M. (2022). Benchmarking for Recommender System (MFRISE). 3C TIC Cuadernos de Desarrollo Aplicados a Las TIC. https://doi.org/10.17993/3ctic.2022.112.146-156

2. Recommendation System. (s/f). NVIDIA Data Science Glossary. Recuperado el 23 de noviembre de 2024, de https://www.nvidia.com/en-us/glossary/recommendation-system/

3. Steck, H., L. Baltrunas, E. Elahi, D. Liang, Y. Raimond, and J. Basilico. 2021. Deep learning for recommender systems: A netflix case study. AI Magazine 42: 718. https://doi.org/10.1609/aaai.12013.

4. B. Smith and G. Linden, "Two Decades of Recommender Systems at Amazon.com," in IEEE Internet Computing, vol. 21, no. 3, pp. 12-18, May-June 2017, doi: 10.1109/MIC.2017.72.

5. Bischof, B., & Yee, H. (2023). Building recommendation systems in python and Jax: Hands-on production systems at scale. OReilly Media.

6. Banik, R. (2018). Hands-On Recommendation Systems with Python: Start building powerful and personalized, recommendation engines with Python. Packt Publishing.

7. Lisandro, K., Shenjun, Z., & Javier, K. (2017). Beating the bookies with their own numbers - and how the online sports betting market is rigged. En arXiv [stat.AP]. http://arxiv.org/abs/1710.02824
